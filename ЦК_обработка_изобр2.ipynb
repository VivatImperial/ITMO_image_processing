```python
file_name = 'sven-hornburg-TAVB-2C6uyE-unsplash.jpg.1'
img = cv2.imread(file_name)
plt.imshow(img)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
h, w, c = RGB_img.shape
new_width = 520

k = float(w) / h
dim = (new_width, int(new_width / k))

resized_RGB_img = cv2.resize(RGB_img, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_RGB_img)
plt.show()
```

```python
print(k)
print(resized_RGB_img.shape)
```

```python
file_name = 'final.jpg'
cv2.imwrite(file_name, cv2.cvtColor(resized_RGB_img, cv2.COLOR_RGB2BGR))
```
